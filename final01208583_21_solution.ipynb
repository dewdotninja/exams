นิยามฟังก์ชัน cpoly() ดังนี้

```python
# define cubic polynomial functgion
def cpoly(q0,qf,t,qd0=0, qdf=0):
    tf = max(t)
    X = np.array([[0.,0.,0.,1.],[tf**3,tf**2,tf,1],[0.,0.,1.,0.],[3*tf**2,2*tf,1.,0.]])
    b = np.array([q0,qf,qd0,qdf]).T
    
    coeffs = np.linalg.inv(X)@b
    # coefficients of derivatives
    coeffs_d = coeffs[0:3]*np.arange(3,0,-1)
    coeffs_dd = coeffs_d[0:2]*np.arange(2,0,-1)
    q = polyval(coeffs,t)
    q_d = polyval(coeffs_d,t)
    q_dd = polyval(coeffs_dd,t)
    return q,q_d,q_dd
    
    
```

```python
# helper function to evaluate polynomial
def polyval(P,X):
    Y = 0*X
    n = len(P)
    for i in range(n):
        Y = Y +P[i]*X**(n-i-1)
    return Y
```

```python
# generate trajectory for each segment.
t1 = np.arange(0,2,0.01)
t2 = np.arange(0,1,0.01)
q11,qd11,qdd11 = cpoly(0,10,t1,0,10)
q12,qd12,qdd12 = cpoly(10,-5,t1,10,-5)
q13,qd13,qdd13 = cpoly(-5,0,t2,-5,0)
```

```python
# join the segments
q1 = np.hstack((q11,q12,q13))
qd1 = np.hstack((qd11,qd12,qd13))
qdd1 = np.hstack((qdd11,qdd12,qdd13))
t = np.arange(0,5,0.01)
```

```python
# Plot the trajectory
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(8,8))
fig.suptitle('3-segment cubic polynomial')
ax1.plot(t,q1,'b-')
ax1.grid(True)
ax1.set_ylabel('Joint angles')

ax2.plot(t,qd1,'b-')
ax2.grid(True)
ax2.set_ylabel('Joint velocity')

ax3.plot(t,qdd1,'r-')
ax3.set_xlabel('time (s)')
ax3.set_ylabel('Joint acceleration')
ax3.grid(True)
```

#### P2 (20 คะแนน)

ออกแบบตัวควบคุมสำหรับข้อต่อหุ่นยนต์ขับเคลื่อนโดยดีซีมอเตอร์ มีฟังก์ชันถ่ายโอนเท่ากับ

\begin{equation} 
P(s) = \frac{1}{zs^2 + 0.01s}
\end{equation}

โดยมีข้อกำหนดที่ต้องการดังนี้
<ol>
<li> ค่าผิดพลาดในสถานะนิ่งเป็นศูนย์ </li>
<li> การรบกวนในย่านความถี่ต่ำถูกลดทอนลง $0.00y$ หรือน้อยกว่า ในย่านความถี่ต่ำกว่า 0.1 rad/s </li>
<li> สัญญาณรบกวนในย่านความถี่สูงถูกลดทอนลง $0.0z$ หรือน้อยกว่า ในย่านความถี่สูงกว่า 80 rad/s </li>
<li> ระบบป้อนกลับเสถียร โดยมีค่าความเผื่อเฟสอย่างน้อย 30 องศา</li>
</ol>

ใช้โค้ดจากการบ้าน HW2 เกณฑ์การให้คะแนนคือ
<ul>
<li />ผลตอบสนองความถี่ของ $L(s)$ ได้ตามข้อกำหนด (10 คะแนน)
<li />จำลองการทำงาน (tracking, disturbance & noise attenuation) ให้เห็นว่าได้ตามข้อกำหนด (10 คะแนน)
</ul>

```python
# From student ID
y = 5
z = 8

S_bnd_abs = 0.001*y
T_bnd_abs = 0.01*z
# compute bounds
S_bnd = 20*np.log10(S_bnd_abs)
L_bnd = -S_bnd
print("Low frequency bound on S = " + str(S_bnd) + " dB")
print("Low frequency bound on L = " + str(L_bnd) + " dB")
T_bnd = 20*np.log10(T_bnd_abs)
print("High frequency bound on T and L = "+str(T_bnd)+" dB")
```

ดังนั้นจากโจทย์จะได้เงื่อนไขสำหรับ loopshaping ดังนี้
<ol>
<li> $L(s)$ ต้องมีตัวอินทิกรัล สังเกตว่ามีอยู่ใน $P(s)$ อยู่แล้วจึงไม่จำเป็นต้องเพิ่มอีก</li>
<li> $|S(j\omega)| \leq 46 dB \rightarrow |L(j\omega)| \geq 46 dB$ ในย่านความถี่ต่ำกว่า 0.1 rad/s</li>
<li> $|T(j\omega)| \leq -22 dB \rightarrow |L(j\omega)| \leq -22 dB$ ในย่านความถี่สูงกว่า 80 rad/s</li>
<li> $L(j\omega)$ มีค่าเผื่อเฟสอย่างน้อย 30 องศา </li>
</ol>

ใช้ฟังก์ชัน lshape() จากการบ้าน HW2

```python
def lshape(C,P, lf, lfb, hf, hfb, pm ):
    assert lf > 0   # avoid bad values
    assert lfb > 0
    assert hf > lf
    assert hfb < 0
    assert 0 < pm < 90
    L = C*P # form loop transfer function
    # create a suitable range of frequency from lf, hf
    lf_log10 = np.log10(lf)
    w_start = np.floor(lf_log10)-1
    hf_log10 = np.log10(hf)
    w_end = np.ceil(hf_log10)+1
    w = np.logspace(w_start,w_end, 1000)
    
    # frequency response of L
    Lmag, Lph, om = ctl.freqresp(L, w)
    Lmag_db = np.squeeze(20*np.log10(Lmag))
    Lph_deg = np.squeeze(np.degrees(Lph))
    
    # create bound vectors
    lf_mask = np.where(om<lf, lfb, 0)
    hf_mask = np.where(om<hf, 0, hfb)
    lf_bnd = lf_mask*np.ones(om.shape)
    hf_bnd = hf_mask*np.ones(om.shape)
    
    # check whether violation occurs
    lf_idxv = np.where(om>lf)
    lf_idx = lf_idxv[0][0]   # find index of low-freq region
    hf_idxv = np.where(om<hf)
    hf_idx = hf_idxv[0][-1]  # find index of high-freq region
    lfmag = Lmag_db[:lf_idx]
    hfmag = Lmag_db[hf_idx:]
    if min(lfmag)<lfb:
        lf_legend = "LF bound (violated)"
    else:
        lf_legend = "LF bound"
    if max(hfmag)>hfb:
        hf_legend = "HF bound (violated)"
    else:
        hf_legend = "HF bound"    
    
    # desired phase margin vectors
    pmvec = (pm-180)*np.ones(om.shape)
    
    # compute gain/phase margins
    g_margin, ph_margin, wgm, wpm = ctl.margin(L)
    ph_at_crossover = (ph_margin-180)
    # Loopshaping plot
    fig, (ax1, ax2) = plt.subplots(2, figsize=(8,8))
    fig.suptitle('$L(j\omega)$ v.s. bounds')
    ax1.semilogx(om, Lmag_db,'k-', om, lf_bnd, 'm-.', om, hf_bnd,'b-.')
    ax1.legend(["$|L(j\omega)|$",lf_legend,hf_legend],loc="lower left")
    ax1.grid(True)
    ax1.set_ylabel('magnitude (dB)')
    
    ax2.semilogx(om, Lph_deg,'k-',om, pmvec,'b-', wpm, ph_at_crossover,'r*')
    if ph_margin > pm:
        pmtext = "phase margin = " + str(round(ph_margin)) + " degrees"
    else:
        pmtext = "phase margin = " + str(round(ph_margin)) + " degrees (violated)"
        
    ax2.text(wpm,ph_at_crossover,pmtext)
    
    dpmtext = "Desired PM (" + str(pm) + " degrees)"
    ax2.set_xlabel('frequency (rad/s)')
    ax2.set_ylabel('phase (deg)')
    ax2.legend(["$\measuredangle L(j\omega)$",dpmtext],loc="lower left")
    ax2.grid(True)    
    
    # plot magnitude of S and T v.s bounds
    S = 1/(1+L)
    T = L/(1+L)
    
    # frequency responses of S and T
    Smag, Sph, om = ctl.freqresp(S, w)
    Tmag, Tph, om = ctl.freqresp(T, w)
    
    Smag_db = np.squeeze(20*np.log10(Smag))
    Tmag_db = np.squeeze(20*np.log10(Tmag))
   
    # check whether violation occurs
    lf_idxv = np.where(om>lf)
    lf_idx = lf_idxv[0][0]   # find index of low-freq region
    hf_idxv = np.where(om<hf)
    hf_idx = hf_idxv[0][-1]  # find index of high-freq region
    lfSmag = Smag_db[:lf_idx]
    hfTmag = Tmag_db[hf_idx:]
    if max(lfSmag)>-lfb:
        lf_legend = "LF bound (violated)"
    else:
        lf_legend = "LF bound"
    if max(hfTmag)>hfb:
        hf_legend = "HF bound (violated)"
    else:
        hf_legend = "HF bound"    
        
    # create data vector for stability bound in mid freq region
    
    om_mid = om[lf_idx:hf_idx]
    pm_r = np.radians(pm)
    x = np.sin(0.5*(np.pi - pm_r))/(np.sin(pm_r)) 
    x = 20*np.log10(x)
    bnds_mid = x*np.ones(om_mid.shape)
    
    ST_peak = max(max(Smag_db),max(Tmag_db))
    if ST_peak > x:
        mf_legend = "Stability bound (violated)"
    else:
        mf_legend = "Stability bound"
    
    plt.figure(figsize=(8,4))
    plt.semilogx(om,Smag_db,'k-',om,Tmag_db,'g-',om,-lf_bnd,'m-.',om,hf_bnd,'b-.',om_mid,bnds_mid,'r-.')
    plt.xlabel('frequency (rad/s)')
    plt.ylabel('magnitude (dB)')
    plt.legend(["$|S(j\omega)|$","$|T(j\omega)|$",lf_legend,hf_legend, mf_legend])
    plt.grid(True)
    plt.title('$|S(j\omega)|$ and $|T(j\omega)|$ v.s. bounds')
    plt.show()
```

```python
# create the plant
s = ctl.tf('s')
P = 1/(z*s**2 + 0.01*s)
P
```

```python
lf = 0.1 # low frequency here
lfb = 46 # your answer 
hf = 80 # high frequency here
hfb = -22 # your answer
pm = 30
```

```python
# adjust gain, zero and pole of C(s) until all the specs are met

C = 4000*(s+0.1)/(s+30) # your controller
lshape(C,P, lf, lfb, hf, hfb,pm )
```

#### จำลองการทำงาน (tracking, disturbance & noise attenuation) ให้เห็นว่าได้ตามข้อกำหนด

```python
L = C*P
T = L/(1+L)
tvec = np.arange(0,1.0,0.01)
r = np.ones(tvec.shape)
tout, y = ctl.step_response(T,tvec)
plt.figure(figsize=(8,4))
plt.plot(tout,r,'b-.',tout,y,'r-')
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('output')
plt.legend(['step ref.','y(t)'])
plt.title('Closed-loop step response')
plt.show()
```

```python
def plot_response(sys,u,t,title):
    tout, y = ctl.forced_response(sys,t, u)
    truncated_idx = 150  # get rid of transient
    tout = tout[truncated_idx:]
    u = u[truncated_idx:]
    y = y[truncated_idx:]
    fig, (ax1, ax2) = plt.subplots(2, figsize=(8,8))
    fig.suptitle(title)
    ax1.plot(tout,u,'b-')
    ax1.grid(True)
    ax1.set_ylabel('input')
    
    ax2.plot(tout,y,'r-')
    ax2.grid(True)
    ax2.set_ylabel('output')
    ax2.set_xlabel('time (sec)')

    plt.show()    
```

```python
L = C*P
S = 1/(1+L)
t = np.arange(0,200,0.01)  # ปรับค่าคาบเวลาให้เห็นอย่างน้อย 1 ไซเคิล
w = 0.1  # rad/s. ต้องเปลี่ยนให้ตรงกับที่โจทย์กำหนด
u = np.sin(w*t)
plot_response(S,u, t, 'Output disturbance response')
```

จะเห็นว่าขนาดเอาต์พุตที่เกิดจากการรบกวน $d(s)$ มีค่าน้อยกว่า $\pm 0.005$ ตามข้อกำหนด

```python
t = np.arange(0,2,0.001) # adjust if needed
w = 80  # rad/s ต้องเปลี่ยนค่านี้ตามที่โจทย์กำหนด
u = np.sin(w*t)
T = L/(1+L)
plot_response(T,u, t, 'Noise response')
```

จะเห็นว่าขนาดเอาต์พุตที่เกิดจากสัญญาณรบกวน $n(s)$ มีค่าน้อยกว่า $\pm 0.08$ ตามข้อกำหนด

$$

<div align="center"><b>รูปที่ 3a) ความสัมพันธ์ระหว่าง $L(j\omega)$ และ $S(j\omega)$</b></div>


สร้างเวกเตอร์ $X$ จากจุด -1 มาที่จุดใดๆ บน $L(j\omega)$ เมื่อหาความสัมพันธ์ของเวกเตอร์ จะได้ว่า
$$
|L(j\omega)| = -1 + |X| 
$$
$$
1 + |L(j\omega)| = |X|
$$
$$
\frac{1}{1+|L(j\omega)|}=|S(j\omega)|=\frac{1}{|X|}
$$
ดังนั้นจะได้ว่า 
$$
|X| \ge 1 \Rightarrow |S(j\omega)| \le 1
$$
ซึ่งก็คือความสัมพันธ์ที่โจทย์ต้องการให้แสดง

#### Solution :
จากโจทย์ พารามิเตอร์ของข้อต่อหุ่นยนต์คือ $J = z = 8, B = 0.01$

สมมุติว่าค่าที่โมเดลได้คือ $J_m$ และ $B_m$ มีค่าผิดพลาดไม่เกิน 10% 


```python
z = 8
J = z
B = 0.01
Jm = (1+0.1*np.random.rand())*J
Bm = (1+0.1*np.random.rand())*B
print("ค่า J ที่โมเดลได้คือ Jm = {}".format(Jm))
print("ค่า B ที่โมเดลได้คือ Bm = {}".format(Bm))
```

สร้างระบบป้อนกลับที่ประกอบด้วยตัวควบคุม $C(s)$ และตัวควบคุมป้อนข้างหน้า โดยพิจารณาว่าฟังก์ชันถ่ายโอนจากอินพุตของตัวควบคุมป้อนข้างหน้าไปยังเอาต์พุต
คือ $P(s)S(s)$

```python
u_ff = Jm*qdd1 + Bm*qd1 # อินพุตที่ได้จากตัวควบคุมป้อนข้างหน้า
t = np.arange(0,5,0.01)  # เวกเตอร์เวลาจากข้อ P1
sys1 = T  # ฟังก์ชันถ่ายโอนวงปิดจากข้อ P2
sys2 = P*S # ฟังก์ชันถ่ายโอนจากอินพุต u_ff
sys = ctl.append(sys1, sys2) 
u = np.vstack((q1,u_ff))
tout, yout = ctl.forced_response(sys,t,u )
```

```python
y1 = np.sum(yout, axis = 0) # เอาต์พุตที่มีการชดเชยจากตัวควบคุมป้อนข้างหน้า
```

```python
# การจำลองโดยไม่ใช้ตัวควบคุมป้อนข้างหน้า
tout, y2 = ctl.forced_response(T,t,q1)
```

```python
plt.figure(figsize=(8,4))
plt.plot(tout,q1,'b-.',tout,y1,'r-') # กรณีใช้ตัวควบคุมป้อนข้างหน้า
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('output')
plt.legend(["q(t)","y(t)"])
plt.title('Step response with feedforward control')

plt.figure(figsize=(8,4))
plt.plot(tout,q1,'b-.',tout,y2,'r-') # กรณีไม่ใช้ตัวควบคุมป้อนข้างหน้า
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('output')
plt.legend(["q(t)","y(t)"])
plt.title('Step response without feedforward control')
plt.show()
```

#### P4 (20 คะแนน)

จากพลานต์ข้อต่อหุ่นยนต์ในข้อ P2 

\begin{equation} 
P(s) = \frac{1}{zs^2 + 0.01s}
\end{equation}

ออกแบบตัวควบคุมป้อนกลับสเตตและตัวอินทิกรัล (state feedback + integrator control) ให้ได้ตามข้อกำหนดดังนี้
1. การพุ่งเกิน < 15%
2. ช่วงเวลาขาขึ้น (rise time) < 0.2 วินาที

จำลองผลตอบสนองขั้นบันไดให้เห็นว่าได้ตามข้อกำหนด (หรือใกล้เคียงที่สุดพร้อมอธิบายเหตุผลหากทำไม่ได้)

<b>Hint : </b> ใช้โค้ดจาก module 5 ใน <a href="https://github.com/dewdotninja/control_python">control_python</a> github


#### Solution :

สร้างพลานต์ตามโจทย์

```python
z = 8
J = z
B = 0.01
A = np.array([[0.,  1.  ],
 [ 0.  , -B/J]])
B = np.array([[0.],
 [ 1/J]])
C = np.array([[ 1.,  0.]])
D = np.array([0.])
Pss = ctl.ss(A,B,C,D)
```

คำนวณตำแหน่งโพลของระบบวงปิดที่ต้องการวาง

```python
zeta = 0.7
wn = 40  # ปรับค่า natural frequency หากไม่ได้ rise time ตามต้องการ
lamda_s = s**2 + 2*zeta*wn*s+wn**2
desired_poles = ctl.zero(lamda_s) # poles are roots of characteristic polynomial
print(desired_poles)
```

หาค่าอัตราขยายของตัวควบคุมป้อนกลับสเตต

```python
K = ctl.acker(A, B, desired_poles)
print(K)
```

และอัตราขยาย DC

```python
Tss = ctl.ss(A-B@K, B, C, D)
ffgain=1/ctl.dcgain(Tss)
print(ffgain)
```

สร้างพลานต์และพล็อตผลตอบสนองโดยใช้โค้ดจาก module 5

```python
Ki = 200000
A_a = np.hstack((A-B@K, -Ki*B))
C_a = np.hstack((C,[[0]]))
A_a = np.vstack((A_a,C_a))
A_a
B_d = np.vstack((B,0))
B_r = np.vstack((np.zeros(B.shape),-1))
B_a = np.hstack((B_d,B_r))
D_a = np.zeros((1,2))
Tss_a = ctl.ss(A_a, B_a, C_a,D_a) #  two-inputs one-output closed-loop system
```

```python
tvec = np.arange(0,2,0.001)
midpts = int(len(tvec)/2)
dist_level = 0.1
r = np.ones(tvec.shape)
d = -dist_level*ffgain*np.ones(tvec.shape)
d[:midpts] = 0
u = np.vstack((d,r))
t, y = ctl.forced_response(Tss_a, tvec, u)
plt.figure(figsize=(8,4))
plt.plot(tvec,r,'k-.',tvec,y.T,'b-')
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('output')
plt.legend(["ref. cmd","y(t)"])
plt.title('Step response of state feeback control with integrator')
plt.show()
```

#### P5. (15 คะแนน)

จากฟังก์ชันถ่ายโอนของตัวควบคุม PID
$$
C(s) = k_p + \frac{k_i}{s} + \frac{Nk_ds}{s+N}
$$

แปลงตัวควบคุมนี้ให้เป็นแบบดีสครีตและเป็นอัลกอริทึมในรูปสมการผลต่าง จำลองผลตอบสนองขั้นบันไดของระบบป้อนกลับโดยใช้พลานต์จากข้อ P2

\begin{equation} 
P(s) = \frac{1}{zs^2 + 0.01s}
\end{equation}

ปรับค่าของ kp, ki,kd ให้ระบบป้อนกลับเสถียรและได้ผลการตามรอยที่มีค่าแตกต่างสถานะนิ่ง (steady state error) เป็นศูนย์ และเวลาเข้าสู่สถานะนิ่ง 
(settling time) ไม่เกิน 1 วินาที (หากไม่ได้ให้ใกล้เคียงที่สุด)

<b>Hint : </b> เซลด้านล่างแสดงโค้ดบางส่วน เติมโค้ดให้สมบูรณ์

```python
z = 8 # your code here
s = ctl.tf('s')
pnum = 1
pden = z*s**2+0.1*s
P = ctl.tf(pnum/pden)

Ts = 0.01

Pd = ctl.sample_system(P, Ts, method='tustin') # convert to discrete-time
print("Discrete-time plant")
print(Pd)
Pdnum, Pdden = ctl.tfdata(Pd)  # extract numerator and denominator of Pd
#print(Pdnum)
#print(Pdden)

# --- code for discrete-time simulation of the plant ---
pdnum = np.squeeze(np.array(Pdnum))
pdden = np.squeeze(np.array(Pdden))

tvec = np.arange(0,2,Ts)  # change time scale to suit
yvec = np.zeros(tvec.shape)
uvec = np.zeros(tvec.shape)
rvec = np.ones(tvec.shape) # step command vector
numorder = len(pdnum)  # 4 for this example
denorder = len(pdden)  # 4
y_prev = np.zeros((denorder,1)) 
u_prev = np.zeros((numorder,1))
# -------------------------------------------------

# ----define PID control variables -------

e1 = 0   # past error input
e0 = 0   # current error input
ui1 = 0   # past integral-term output 
ui0 = 0   # current integral-term output
up = 0    # proportional term output
ud1 = 0  # past derivative-term output
ud0 = 0  # current derivative-term output

# you must adjust kp, ki,kd to give stable step response with desired tracking performance
kp = 200   # proportional gain. 
ki = 10   # integral gain
kd = 150   # derivative gain

N = 20  # you may change this if desired

# ----- compute coefficients for PID algorithm -----
bi = 0.5*Ts*ki
ad1 = (1 + 0.5*N*Ts)
ad2 = (0.5*N*Ts -1)
ad = -ad2/ad1
bd = kd*N/ad1

# ------------------------

i = 0
# perform discrete-time simulation 
for i in range(len(tvec)):

    for k in range(numorder-1):
        u_prev[k] = u_prev[k+1]
    for k in range(denorder-1):
        y_prev[k] = y_prev[k+1]
        

#  ----- put your PID controller simulation here -----
    e1 = e0    # keep previous loop error
    ui1 = ui0   # keep previous loop integral output
    ud1 = ud0  # keep previous looop derivative output
    e0 = rvec[i] - y_prev[denorder-1] # compute new error
    up =  kp*e0 # proportional-term output
    ui0 =  ui1 + bi*(e0+e1) # integral-term output 
    ud0 =  ad*ud1 + bd*(e0 - e1) # derivative-term output
    uvec[i] = up + ui0 + ud0
# -------------------------------------------------------    
        
    u_prev[numorder-1] = uvec[i]        
    yvec[i] = pdnum[0]*uvec[i]
    for k in range(denorder-1):
        yvec[i] = yvec[i]+pdnum[k+1]*u_prev[numorder-2-k] - pdden[k+1]*y_prev[denorder-2-k]
        #print(pdnum[k+1])
        #print(pdden[k+1])
        #print()
    y_prev[denorder-1] = yvec[i]
    
# plot the result

plt.figure()
plt.plot(tvec, rvec,'m-',tvec, yvec,'b-')
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('y(k)')
plt.legend(["Ref. cmd","Plant output"])

plt.title("Step response")

plt.figure()
plt.plot(tvec,uvec,'r-')
plt.grid('True')
plt.xlabel('time (sec)')
plt.ylabel('u(k)')
plt.title("controller output")



plt.show()
```
